### 실행시간 측정하기

```python
%%time
# 셀 전체 실행시간 측정하기. 꼭 첫줄에 넣어줘야함. comment도 앞서서는 안됨

for i in range(100000):
    12**10
```

```python
%%timeit -n 500 -r 10 # 셀 평균 실행시간 측정. 500번씩 10회 총 5000번 돌린 후 평균

for i in range(100):
    12**10
```

```python
#  %하나 적으면 inline 실행시간 평균 측정
%time 1+1
```
